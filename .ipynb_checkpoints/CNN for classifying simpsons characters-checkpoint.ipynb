```python
import time

import torch
import torchvision
import torchvision.transforms as transforms
```

```python
train_path = '/home/ds/dl/Matt/datasets/simpsons/the-simpsons-characters-dataset/split/train'
test_path = '/home/ds/dl/Matt/datasets/simpsons/the-simpsons-characters-dataset/split/test'
```

### Load Data

```python
img_size = 96
batch_size = 30
```

```python
transform = transforms.Compose(
    [transforms.Resize((img_size, img_size)),
     transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

trainset = torchvision.datasets.ImageFolder(root=train_path, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=3)

testset = torchvision.datasets.ImageFolder(root=test_path, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=False, num_workers=3)
```

### Define NN

```python
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5) # in_channels, out_channels, kernal_size
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 21 * 21, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        #print(x.size())
        x = self.pool(F.relu(self.conv1(x)))
        #print(x.size())
        x = self.pool(F.relu(self.conv2(x)))
        #print(x.size())
        x = x.view(-1, 16 * 21 * 21)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x


net = Net()
```

### Define loss function and optimizer

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

### Train the network

```python
def calculate_accuracy(outputs, labels):
    _, predicted = torch.max(outputs, 1)
    return sum((predicted == labels).squeeze()).item()
```

```python
print_tick = int(len(trainset) / batch_size / 50)
```

```python
f'Starting: {time.ctime()}'

for epoch in range(50):  # loop over the dataset multiple times
    print(f'epoch: {epoch}\t{time.ctime()}')
    running_correct = 0
    running_total = 0
    running_loss = 0.0
    
    for i, data in enumerate(trainloader, 0):
        # get the inputs
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # update accuracy
        running_correct += calculate_accuracy(outputs, labels)
        running_total += len(labels)
        
        # update loss
        running_loss += loss.item()
        
#         if i % 1000 == 999:    # print every n mini-batches
#             print('[%d, %5d] loss: %.3f\t acccuracy: %.3f' %
#                   (epoch + 1, i + 1, running_loss / 2000, running_correct / running_total))
#             running_loss = 0.0
        if i % print_tick == 0:
            print('.', end='')

    print('\n[%d, %5d] loss: %.3f\t acccuracy: %.3f' %
          (epoch + 1, i + 1, running_loss / running_total, running_correct / running_total))



print('Finished Training')
```

```python
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print('Accuracy of the network on the 10000 test images: %d %%' % (
    100 * correct / total))
```

```python
class_correct = list(0. for i in range(10))
class_total = list(0. for i in range(10))
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = net(images)
        _, predicted = torch.max(outputs, 1)
        c = (predicted == labels).squeeze()
        for i in range(min(batch_size, len(labels))):
            label = labels[i]
            class_correct[label] += c[i].item()
            class_total[label] += 1
```

```python
for i in range(10):
    print('Accuracy of %5s : %2d %%' % (
        testset.classes[i], 100 * class_correct[i] / class_total[i]))
```
